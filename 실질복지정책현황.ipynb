```python
where_list=['세종','강원','경남','광주']
where_all_list = policy_df['지역'].value_counts().to_dict().keys()

for 지역 in where_all_list:
    df = policy_df[policy_df['지역']==지역]
    fig = plt.figure(figsize=(15,7))
    p = sns.countplot(x='분야', data=df)
    p.set_title(지역+' 정책분야 현황')
```
